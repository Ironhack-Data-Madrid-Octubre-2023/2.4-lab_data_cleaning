```python
def limpiar_auto(x):  #Con esta fución limpiamos lo siguiente. En la tabla de arriba 👆 vemos que hay: 
                      # 'auto', 'automatic', manual. La función de abajo lo limpiaria.
    
                        
    x = x.lower()      #x, todas las letras, a mínuscula. 
    
    if 'auto' in x:    #si x está en auto, devuelveme automatic.
        return 'automatic'
    
    elif 'man' in x:   # si x está en man, devuelveme manual.
        return 'manual'
    
    else:               #todo lo demás, devuelveme 'unknown'
        return 'unknown'
```

```python
data.trany = data.trany.apply(limpiar_auto)   

data.head()
```
